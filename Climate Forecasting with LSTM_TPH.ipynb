```python
# ================================
# Climate Forecasting with LSTM
# ================================

# 1. Imports
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

# 2. Load Dataset
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)

csv_path = os.path.join(os.path.dirname(zip_path), "jena_climate_2009_2016.csv")
df = pd.read_csv(csv_path)

# Downsample to hourly data
df = df[5::6]
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')

# 3. Helper Function
def prepare_data(df, target_col, window_size=5, scaler=None):
    series = df[target_col].values.reshape(-1,1)
    if scaler is None:
        scaler = MinMaxScaler()
    series_scaled = scaler.fit_transform(series)

    x, y = [], []
    for i in range(len(series_scaled) - window_size):
        x.append(series_scaled[i:i+window_size])
        y.append(series_scaled[i+window_size])
    x, y = np.array(x), np.array(y)

    train_size = int(len(x) * 0.85)
    val_size   = int(len(x) * 0.05)

    x_train, y_train = x[:train_size], y[:train_size]
    x_val, y_val     = x[train_size:train_size+val_size], y[train_size:train_size+val_size]
    x_test, y_test   = x[train_size+val_size:], y[train_size+val_size:]

    y_train, y_val, y_test = y_train.squeeze(), y_val.squeeze(), y_test.squeeze()
    return x_train, y_train, x_val, y_val, x_test, y_test, scaler

# 4. Model Builder
def build_model(input_shape, output_dim=1):
    model = Sequential([
        InputLayer(input_shape),
        LSTM(64),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dense(output_dim)
    ])
    model.compile(loss=MeanSquaredError(),
                  optimizer=Adam(learning_rate=0.0005),
                  metrics=[RootMeanSquaredError()])
    return model

# 5. Training & Visualization Function
def train_and_plot(df, target_col, window_size=5, multivariate=False):
    print(f"\n=== Forecasting {target_col} ===")
    if not multivariate:
        x_train, y_train, x_val, y_val, x_test, y_test, scaler = prepare_data(df, target_col, window_size)
        model = build_model((window_size,1), output_dim=1)
    else:
        # Example: wind speed + direction
        wind = df[['wv (m/s)', 'wd (deg)']].values
        scaler = MinMaxScaler()
        wind_scaled = scaler.fit_transform(wind)

        x, y = [], []
        for i in range(len(wind_scaled) - window_size):
            x.append(wind_scaled[i:i+window_size])
            y.append(wind_scaled[i+window_size])  # both speed & direction
        x, y = np.array(x), np.array(y)

        train_size = int(len(x) * 0.85)
        val_size   = int(len(x) * 0.05)

        x_train, y_train = x[:train_size], y[:train_size]
        x_val, y_val     = x[train_size:train_size+val_size], y[train_size:train_size+val_size]
        x_test, y_test   = x[train_size+val_size:], y[train_size+val_size:]

        model = build_model((window_size,2), output_dim=2)

    # Train
    history = model.fit(x_train, y_train,
                        validation_data=(x_val, y_val),
                        epochs=15,
                        batch_size=32,
                        verbose=1)

    # Evaluate
    test_loss, test_rmse = model.evaluate(x_test, y_test, verbose=1)
    print(f"Test RMSE: {test_rmse:.3f}")

    # Predictions
    preds = model.predict(x_test)
    if not multivariate:
        preds_rescaled = scaler.inverse_transform(preds.reshape(-1,1)).flatten()
        actuals_rescaled = scaler.inverse_transform(y_test.reshape(-1,1)).flatten()

        plt.figure(figsize=(12,6))
        plt.plot(actuals_rescaled[:200], label=f'Actual {target_col}')
        plt.plot(preds_rescaled[:200], label=f'Predicted {target_col}')
        plt.title(f"{target_col} Forecasting: Actual vs Predicted")
        plt.xlabel("Time Steps")
        plt.ylabel(target_col)
        plt.legend()
        plt.show()
    else:
        # Plot speed only for simplicity
        preds_rescaled = scaler.inverse_transform(preds)[:,0]
        actuals_rescaled = scaler.inverse_transform(y_test)[:,0]

        plt.figure(figsize=(12,6))
        plt.plot(actuals_rescaled[:200], label='Actual Wind Speed')
        plt.plot(preds_rescaled[:200], label='Predicted Wind Speed')
        plt.title("Wind Speed Forecasting: Actual vs Predicted")
        plt.xlabel("Time Steps")
        plt.ylabel("Wind Speed (m/s)")
        plt.legend()
        plt.show()

# 6. Run Forecasts
train_and_plot(df, 'T (degC)')       # Temperature
train_and_plot(df, 'rh (%)')         # Humidity
train_and_plot(df, 'p (mbar)')       # Pressure
train_and_plot(df, ['wv (m/s)','wd (deg)'], multivariate=True)  # Wind speed + direction
```

```python
# ================================
# Climate Estimation Dashboard (Year + Month + Week)
# ================================

import dash
from dash import dcc, html, Input, Output
import pandas as pd
import numpy as np

# Assume df is already loaded and downsampled to hourly data with datetime index
# df = df[5::6]
# df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')

# Initialize Dash app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Climate Estimation Dashboard"),

    html.Label("Select Year:"),
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(y), 'value': y} for y in df.index.year.unique()],
        value=df.index.year.min()
    ),

    html.Label("Select Month:"),
    dcc.Dropdown(
        id='month-dropdown',
        options=[{'label': pd.to_datetime(f'2020-{m}-01').strftime('%B'), 'value': m} for m in range(1,13)],
        value=1
    ),

    html.Label("Select Week:"),
    dcc.Dropdown(
        id='week-dropdown',
        options=[{'label': f'Week {w}', 'value': w} for w in range(1,5)],  # approx 4 weeks per month
        value=1
    ),

    html.Div(id='output-div')
])

# Callback
@app.callback(
    Output('output-div', 'children'),
    [Input('year-dropdown', 'value'),
     Input('month-dropdown', 'value'),
     Input('week-dropdown', 'value')]
)
def update_output(year, month, week):
    # Define week ranges (approx: 7-day blocks)
    start_day = (week-1)*7 + 1
    end_day = start_day + 6

    # Filter data
    mask = (df.index.year == year) & (df.index.month == month) & (df.index.day >= start_day) & (df.index.day <= end_day)
    subset = df.loc[mask]

    if subset.empty:
        return "No data available for this selection."

    # Min & Max Temperature
    min_temp = subset['T (degC)'].min()
    max_temp = subset['T (degC)'].max()

    # Detect abnormal spikes (> 5°C change day-to-day)
    daily_avg = subset['T (degC)'].resample('D').mean()
    spikes = []
    for i in range(1, len(daily_avg)):
        if abs(daily_avg.iloc[i] - daily_avg.iloc[i-1]) > 5:
            spikes.append(daily_avg.index[i].strftime('%Y-%m-%d'))

    spike_msg = "No abnormal spikes detected."
    if spikes:
        spike_msg = f"Abnormal spikes detected around: {', '.join(spikes)}"

    # Humidity analysis
    avg_hum = subset['rh (%)'].mean()
    if avg_hum > 70:
        suggestion = "High humidity — carry essentials like umbrella or raincoat."
    elif avg_hum < 40:
        suggestion = "Low humidity — sunny/dry conditions, stay hydrated and carry an umbrella."
    else:
        suggestion = "Moderate humidity — normal conditions, light precautions recommended."

    return html.Div([
        html.H3(f"Year {year}, Month {pd.to_datetime(f'2020-{month}-01').strftime('%B')}, Week {week}"),
        html.P(f"Min Temperature: {min_temp:.2f} °C"),
        html.P(f"Max Temperature: {max_temp:.2f} °C"),
        html.P(spike_msg),
        html.P(f"Average Humidity: {avg_hum:.2f}%"),
        html.P(f"Suggestion: {suggestion}")
    ])

if __name__ == '__main__':
    app.run(port=8050, debug=True)
```
